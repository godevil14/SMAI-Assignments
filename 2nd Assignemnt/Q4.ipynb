```python
filename = "/content/drive/My Drive/SMAI_A2_2024/training_3digits.hdf5"
train = h5py.File(filename,'r')
train_images = np.array(train['images'])
train_digits = np.array(train['digits'])
train.close()
```

```python
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score

train_images = train_images.reshape(train_images.shape[0], -1)

mlp_clf = MLPClassifier(hidden_layer_sizes=(100,), max_iter=100, alpha=1e-4,
                        solver='adam', verbose=10, random_state=1,
                        learning_rate_init=.1)
mlp_clf.fit(train_images, train_digits)
```

```python
test_filename_1 = "/content/drive/My Drive/SMAI_A2_2024/testing_3digits_part1.hdf5"
test1 = h5py.File(test_filename_1, 'r')
test_images_1 = np.array(test1['images'])
test_digits_1 = np.array(test1['digits'])
test1.close()
test_images_1 = test_images_1.reshape(test_images_1.shape[0], -1)
```

```python
predictions_part1 = mlp_clf.predict(test_images_1)
accuracy_part1 = accuracy_score(test_digits_1, predictions_part1)
print("Accuracy for test set part 1:", accuracy_part1)
```

```python
filename = "/content/drive/My Drive/SMAI_A2_2024/testing_3digits_part2.hdf5"
test2 = h5py.File(filename,'r')
test_images_2 = np.array(test2['images'])
test_digits_2 = np.array(test2['digits'])
test2.close()

test_images_2 = test_images_2.reshape(test_images_2.shape[0], -1)

predictions_part2 = mlp_clf.predict(test_images_2)
accuracy_part2 = accuracy_score(test_digits_2, predictions_part2)
print("Accuracy for test set part 2:", accuracy_part2)
```
